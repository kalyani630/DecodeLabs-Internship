```python
import pandas as pd
```

```python
df=pd.read_excel("Cleaned Dataset for Data Analytics.xlsx")
df
```

```python
#Display the dataset information
df.info()

#dataset shape
print(f"Dataset Shape: {df.shape}")
```

```python
#Statistics
df.describe()
```

```python
numerical_cols=df.select_dtypes(include=['int64','float64']).columns
print(numerical_cols)
```

```python
print("Mean:")
print(df[numerical_cols].mean())
```

```python
print("Count:")
print(df.count())
```

```python
print("Median:")
print(df[numerical_cols].median())
```

```python
categorical_cols=df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```

```python
print("\nOutlier Detection:")
Q1 = df['TotalPrice'].quantile(0.25)
Q3 = df['TotalPrice'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['TotalPrice'] < lower_bound) | (df['TotalPrice'] > upper_bound)]
print(f"{'TotalPrice'}: {len(outliers)} outliers")
print(outliers[['Product','Quantity','TotalPrice']])
```

```python
# Calculate total quantity sold for each product
product_sales = df.groupby('Product')['Quantity'].sum().reset_index()

# Display product-wise sales quantity
print(product_sales)
```

```python
# Calculate daily sales trend
daily_sales = df.groupby('Date')['TotalPrice'].sum()

# Display daily sales trend
print("Daily Sales Trend:\n",daily_sales)
```

```python
# Extract month from Date
df['Month'] = df['Date'].dt.to_period('M')

# Calculate monthly sales
monthly_sales = df.groupby('Month')['TotalPrice'].sum()

# Display monthly sales
print(monthly_sales)
```

```python
#Count orders per month
monthly_orders = df.groupby('Month')['OrderID'].count()
#Display monthly order counts
print(monthly_orders)
```

```python
# Payment Method Distribution
payment_method = df['PaymentMethod'].value_counts()

print(payment_method)
```

```python
# Order Status Distribution
order_status = df['OrderStatus'].value_counts()

print(order_status)
```
